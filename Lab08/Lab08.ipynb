## Задание 1
Выяснить, образуют ли группы следующие множества при указанной операции над элементами:

8) матрицы порядка n с целыми элементами и определителем, равным ±1 относительно умножения.

### Решение

Множество образует группу, если выполняются следующие условия:
1. Ассоциативность
2. Наличие нейтрального элемента
3. Наличие обратного элемента

Проверим выполнение условий (1) - (3):

1. $$
 (a * b)* c = (a \times b)\times c = a\times b\times c = a \times (b\times c) = a * (b * c).
 $$
 По свойству определителя произведения:
 $$
 |(a \times b)\times c| = (|a| \times |b|)\times |c| = |a|\times |b|\times |c| = \pm 1
 $$

2. Можно показать, что нейтральным элементом является единичная матрица порядка $n$:
 $$
 a* e = a\times e = e\times a = e * a = a
 $$
 Определитель произведения:
 $$
 |a|\times |e| = |a|\times 1 = |a|.
 $$

3. Так как $|A^{-1}|=\frac{1}{|A|}$ для невырожденной матрицы, то для любой матрицы $A$, такой что $|A|=\pm 1$, будет существовать обратный элемент $A^{-1}$ c определителем $|A^{-1}|=\frac{1}{|A|}=\pm 1$

Так как все уловия выполняются, множество образует группу.

## Задание 4
Составить программу, которая для любой подстановки порядка $n$ определяет она четная или нечётная.

### Решение

Для решения задачи будем определять количество инверсий в подстановке, по определению, если их количество четно, то подстановка четная.

```python
# определение количества инверсий в подстановке
def inversions(s): 
    count = 0
    for i in range(len(s)): 
        for j in range(i + 1, len(s)): 
            if (s[i] > s[j]): 
                count += 1
    return count
```

Для создания интерфейса определим вспомогательные функции:

```python
# конверсия строки в подстановку 
def parse(s):
    res = map(int, filter(None, s.split(' ')))
    return list(res)
```

```python
# форматированный вывод подстановки
def print_sub(sub):
    identity = list(range(1, len(sub) + 1))
    res = f"{' & '.join([str(i) for i in identity])} \\\\ { '& '.join([str(i) for i in sub])}"
    return r'$\begin{pmatrix}'+res+r'\end{pmatrix}$'
```

```python
import ipywidgets as widgets
from ipywidgets import interact
from IPython.display import Latex

@interact(Подстановка = "2 1 3")
def is_even(Подстановка):
    
    sub = parse(Подстановка)
    inv = inversions(sub)
    res = 'четная' if inv % 2 == 0 else 'нечетная'
    out = rf'Подстановка {print_sub(sub)} {res} ({inv} инверсий)'
    return Latex(out)
```
